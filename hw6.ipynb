```python
# MTH 451 HW 6 (Project 2)
# Jackson Lyons
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
# Order of Convergence for Composite Trapezoidal Rule

f = lambda x: np.exp(x)
a, b = 0, 4
h = (b-a)
# n = # subintervals
n = 1
hs = np.linspace(0, h, 20)
h_vals = []
errors = []

# Define error bound function O(h^2)
bound = lambda h: h**2

# Exact value of integral
val_exact = f(4)-f(0)

for i in range(0, 10):
  print("h=", h)
  print("# subintervals=", n)
  x = np.arange(a, b+h, h)
  y = f(x)
  val_trapz = integrate.trapz(y, x)
  abs_error = abs(val_exact - val_trapz)
  print("val=", val_trapz, "abs_error=", abs_error)
  print("err/h^2", abs_error/h**2, '\n')
  h_vals.append(h)
  errors.append(abs_error)
  n*=2
  h = (b-a)/n

# Plot error bound
fig, axs = plt.subplots(2, 1, constrained_layout=True, figsize=(6,8))
axs[0].plot(h_vals, errors, '-o', label='e')
axs[0].plot(hs, bound(hs), label='h^2')
axs[0].set_title('h vs. error')
axs[0].set_xlabel(r"$h$", fontsize=14)
axs[0].set_ylabel(r"$error$", fontsize=14)
axs[0].legend(loc="upper left")
fig.suptitle('Convergence of Composite Trapezoidal Rule', fontsize=16)

m, b = np.polyfit(np.log(h_vals), np.log(errors), 1)
axs[1].loglog(h_vals, errors, '-o', label=f'e(h)={m}h + {b}')
axs[1].set_title('log(h) vs. log(error)')
axs[1].set_xlabel(r"$log(h)$", fontsize=14)
axs[1].set_ylabel(r"$log(error)$", fontsize=14)
axs[1].legend(loc="upper left")


print('slope=', m)
```

In the above data and figures, observe that after each iteration where the step size (h) approaches 0 (h->0) and n (number of subintervals) increases, the result of err/h^2 converges to ~4.466 (non-zero constant). This is evidence that the composite simpsons rule has a convergence rate of O(h^2). Another way to illustrate the convergence is the slope of the log(error) vs. log(h). From the second graph, this slope is nearly 2.


```python
# Order of Convergence for Composite Simpson's Rule

f = lambda x: np.exp(x)
a, b = 0, 4
h = (b-a)/2
# n= # subintervals
n=2
hs = np.linspace(0, h, 20)
h_vals = []
errors = []

# Define error bound function O(h^4)
bound1 = lambda h: h**4

# Exact value of integral
val_exact = f(4)-f(0)

for i in range(0, 10):
  print("h=", h)
  print("# subintervals=", n)
  x = np.arange(a, b+h, h)
  y = f(x)
  val_simps = integrate.simps(y, x)
  abs_error = abs(val_exact - val_simps)
  print("val=", val_simps, "abs_error=", abs_error)
  print("err/h^4=", abs_error/h**4, '\n')
  h_vals.append(h)
  errors.append(abs_error)
  n*=2
  h/=2

# Plot error
fig, axs = plt.subplots(2, 1, constrained_layout=True, figsize=(6,8))
axs[0].plot(h_vals, errors, '-o', label='e')
# We know that the error of composite Simpson's Rule is O(h^4)
# Plot error bound h^4
axs[0].plot(hs, bound1(hs), label='h^4')
axs[0].set_title('h vs. error')
axs[0].set_xlabel(r"$h$", fontsize=14)
axs[0].set_ylabel(r"$error$", fontsize=14)
axs[0].legend(loc="upper left")
fig.suptitle('Convergence of Composite Simpsons Rule', fontsize=16)

# fit line to log(h) vs log(error)
m, b= np.polyfit(np.log(h_vals), np.log(errors), 1)
axs[1].loglog(h_vals, errors, '-o', label=f"e(h) = {m}h + {b} ")
axs[1].set_title('log(h) vs. log(error)')
axs[1].set_xlabel(r"$log(h)$", fontsize=14)
axs[1].set_ylabel(r"$log(error)$", fontsize=14)
axs[1].legend(loc="upper left")


print('slope=', m)
```

In the above data and figures, observe that after each iteration where the step size (h) approaches 0 (h->0) and n (number of subintervals) increases, the result of err/h^4 converges to ~0.297 (non-zero constant). This is evidence that the composite simpsons rule has a convergence rate of O(h^4). Another way to illustrate the convergence is the slope of the log(error) vs. log(h). From the second graph, this slope is nearly 4.
